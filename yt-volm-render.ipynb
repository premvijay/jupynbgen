```python
ds = yt.load("/mnt/home/student/cprem/Downloads/IsolatedGalaxy/galaxy0030/galaxy0030")
```

```python
from yt.visualization.volume_rendering.api import Scene, Camera, create_volume_source
```

```python
sc = Scene()
```

```python
source = create_volume_source(ds.all_data(), 'density')
```

```python
sc.add_source(source)
cam = sc.add_camera(ds)
```

```python
cam.position = np.array([0.5, 0.5, 1])
cam.focus = np.array([0.5, 0.5, 0.5])
cam.north_vector = np.array([0.0, 1.0,0.1])
sc.render()
sc.show()
```
